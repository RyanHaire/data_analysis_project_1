# Google Play Store Analysis

## Initial Thoughts Organization 

- Organize apps into respective categories
- Calculate max and min rating of corresponding categories
- Display findings in bar graph

```python
# get unique categories from the category column using set()
app_categories = set(data['Category'])
app_categories
```

```python
# notice the first item in categories is the value of '1.9', 
# we need to remove it since it is not a category
app_categories.remove('1.9')
app_categories
```

```python
# displays quick summary of the data
data['Rating'].describe()
```

```python
# from above we notice that max rating is 19
# it is above the max rating of 5 with any given app in the google play store
# so we have to search and destroy the row with rating of 19 since i am using rating for most of my analysis
```

```python
#  create dictionary to hold keys of categories 
# and the rows corresponding to the category
category_rows = {}
# loop through categories and find rows with category equal to the category
for category in app_categories:
# create category key in category_rows and append rows found from data with category as the current category
# using DataFrame.loc()
# DataFrame.loc() returns a DataFrame object
    category_rows[category] = data.loc[data['Category'] == category]

category_rows
```

```python
category_rows['EDUCATION']
```

```python
y = []
x = []
for key, value in category_rows.items():
    x.append(key)
    y.append(value.shape[0])
    
```

```python
plt.bar(x, y)
plt.xlabel('App Categories')
plt.ylabel('# of Apps')
plt.style.use('fivethirtyeight')
plt.xticks(rotation=90)
# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Set figure width to 25 and height to 9
fig_size[0] = 25
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size


plt.show()
```
